## Import Libraries

```python
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

## Load Data

```python
file = '../data/sampledata.csv'
df_data = pd.read_csv(file)
```

```python
num_cols = len(df_data.columns)
num_rows = len(df_data)
feature_names = df_data.columns[:num_cols-1].values.tolist()

print ("- Number of features in data:", num_cols-1)
print ("- Features: %s" % ", ".join(feature_names))
print ("- Number of rows in data:", num_rows)
```

## Analyse Data
### (1) Data Distribution

```python
def histogram(df_data):
    # convert DataFrame to numpy array
    data = df_data.values

    # define figure grid
    if num_cols < 6:
        row, col = 1, num_cols
    else:
        row, col = num_cols//5+1, 5

    # create figure object
    fig = plt.figure(figsize = (col*5, row*3))
    
    # plot data
    for i in range(num_cols):
        plt.subplot (row, col, i+1)
        plt.title (df_data.columns[i])
        sns.distplot(df_data[df_data.columns[i]].values)
```

```python
histogram(df_data)        
```

### (2) Correlation Coefficient

```python
# heatmap of the correlation coefficient matrix
cr = np.corrcoef (df_data.values.T)
fig = plt.figure()

hm = sns.heatmap(cr, cbar = True, annot = True, square = True, 
                 fmt = '.2f', annot_kws = {'size': 10},
                 yticklabels = feature_names + ["Label"],
                 xticklabels = feature_names + ['Label'])
plt.show ()
```
